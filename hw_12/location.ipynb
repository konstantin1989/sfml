```python
import reverse_geocoder as rg

def gen_coordinates_features(df):
    # Временный признак с кортежем координат
    df['coordinates'] = df['location'].map(lambda x: (x['latitude'], x['longitude']))
    # Временный признак с результатом работы reverse_geocoder
    df['rg_result'] = rg.search(df['coordinates'].tolist())
    
    # Нужны ли фичи с широтой и долготой?
    df['latitude'] = df['location'].map(lambda x: x['latitude'])
    df['longitude'] = df['location'].map(lambda x: x['longitude'])
    
    # Забираем из созданного признака все, что может пригодиться.
    # Буквенный код страны:
    df['rg_country_code'] = df['rg_result'].map(lambda x: x['cc'])
    # Название населенного пункта (тут, судя по всему, вплоть до названия села, не уверен, что может пригодиться):
#     df['rg_city_name'] = df['rg_result'].map(lambda x: x['name'])
    # Насколько я понял, название более крупного образования (область?), 
    # в состав которого входит текущий населенный пункт, вот его можно попробовать использовать
    df['rg_admin1'] = df['rg_result'].map(lambda x: x['admin1'])  
    # И последнее значение, редко заполнено, вряд ли пригодится
#     df['rg_admin2'] = df['rg_result'].map(lambda x: x['admin2'])  
    
    # Дропаем временные столбцы
    df = df.drop(['coordinates', 'rg_result'], axis=1)
    
    return df
```

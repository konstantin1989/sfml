```python
#достаем категорию и подкатегорию из дерева, что приходит в "fields"
def parse_field_category(spisok):
    d = ast.literal_eval(spisok) 
    for i in d:
        if len(i) > 1: 
            if i['field']['type'] == 'category':
                return i['field']['name']
        else:
            return 'Прочее'
            
            
def parse_field_subcategory(spisok):
        d = ast.literal_eval(spisok) 
        for i in d:
            if len(i) > 1: 
                if i['field']['type'] == 'subcategory':
                    return i['field']['name']
            else:
                 return 'Прочее'
```

```python
df2['fields'][0:10].map(parse_field_category)
```

```python
df2['cat_name'] = df2['fields'].map(parse_field_category)
```

```python
df2['subcat_name'] = df2['fields'].map(parse_field_subcategory)
```

```python
def cross_valid(matrix_counts, target, test_size=0.3, random_state=0):
    X_train, X_test, Y_train, Y_test = train_test_split(matrix_counts, target, test_size=0.3, random_state=0)
    return (X_train, X_test, Y_train, Y_test)
```

```python
import numpy as np
```

```python
def rlmse_score(y_test, y_hat):
    rlmse = np.sqrt(np.mean((np.log(y_test + 1) - np.log(y_hat + 1))**2))
    
    return rlmse
```
